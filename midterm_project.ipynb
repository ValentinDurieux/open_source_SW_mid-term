# Load Data Points (Do not modify the following block)

```python
with open('training_data.npz', 'rb') as f:
    data = np.load(f)

    x_list = data['x_list']
    y_list = data['y_list']

    x_data = data['x_data']
    y_data = data['y_data']
    n_data = len(x_data)
    
    w = data['w']
    original_degree = data['order']
    
# Print information of original function.
print("=================================")
print("We have", n_data, "number of data")

print("=================================")
weight_info_string = ''
for d in range(original_degree):
    weight_info_string += 'w'+str(d)+':'+str(round(w[d],ndigits=3))+'  '
print("Coefficients of the original polynomial")
print(weight_info_string)
print("=================================")
    
plt.plot(x_list, y_list, 'b:', linewidth=2, label="Original Function")
plt.scatter(x_data, y_data, s=50, c='r', label="Data Points")
plt.xlim([np.min(x_list),np.max(x_list)])
plt.ylim([np.min(y_data),np.max(y_data)])
plt.legend(prop={'size': 12})
plt.title("Data Plot")
plt.show()
```

### Our goal is to estimate 'w' from data points, 'x_data' and 'y_data'. Answer the following problems.

### 1. Compute a Vandermonde matrix when the degree of polynomial is $4$ (30pt)
- The variable 'degree' is the order of polynomial. In this problem, we set degree=$4$
- Use the variable 'A' for the Vandermonde matrix. Now, 'A' is initialized as a zero matrix whose elements are all zero. Fill in the element of the Vandermonde matrix by using power operator (\*\*), for loop, and np.concatenation.

```python
degree = 4
A = np.zeros((n_data, degree+1)) # Dummy initialization

for i in range(n_data):
    for j in range(degree + 1):
        A[i][j] = x_data[i] ** j
```

### Print results (do not modify the following block)

```python
print(A)
```

### 2. Compute the coefficients of polynomial regression using a $4$ degree polynomial  (40pt)
- Use the variable 'degree' and the Vandermonde matrix 'A' in Problem 1.
- The variable 'w_est' is the coefficients of polynomial regression. Now, 'w_est' is initialized as a zero vector. Compute the 'w_est' from 'A' and 'y'
- The variable 'y_est' is an estimated function value corresponding to the input points 'x_list'. Now, it is a zero list and fill the list by computing the estimated function values. In other words, y_est = $\hat{f}($x_list$)$

```python
# w_est = np.zeros((degree+1,1))
w_est = np.linalg.lstsq(A, y_data, rcond=None)[0]
y_est = np.zeros_like(x_list)


for i, x in enumerate(x_list):
    for j in range(degree + 1):
        y_est[i] += w_est[j] * (x ** j)
```

### Print results (do not modify the following block)

```python
plt.plot(x_list, y_list, 'b:', linewidth=2, label="Original Function")
plt.plot(x_list, y_est, 'm-', linewidth=2, label="Polynomial Regression (d={})".format(degree))
plt.scatter(x_data, y_data, s=50, c='r', label="Data Points")
plt.xlim([np.min(x_list),np.max(x_list)])
plt.ylim([np.min(y_data),np.max(y_data)])
plt.legend(prop={'size': 12})
plt.title("Data Plot")
plt.show()
```

The effect of the degree of a polynomial in polynomial regression is an essential aspect to consider when fitting models to data. Let's discuss the pros and cons of both high and low-degree polynomials, and the phenomenon related to this in machine learning, which is known as "bias-variance trade-off."

**High Degree Polynomials:**
Pros:
1. **Flexibility**: High-degree polynomials can closely fit the training data, capturing complex and intricate relationships between the predictors and the target variable. They provide a more flexible model to approximate the data.

2. **Low Bias**: High-degree polynomials have low bias, meaning they are capable of fitting the training data very well, resulting in a model that can capture fine details.

Cons:
1. **Overfitting**: The major drawback of high-degree polynomials is overfitting. When the degree is too high, the model may become too sensitive to noise in the data. It fits the training data perfectly but does not generalize well to unseen data.

2. **Instability**: High-degree polynomials can lead to numerically unstable solutions. Small changes in the input data can result in large variations in the model parameters, making it difficult to interpret and maintain model stability.

**Low Degree Polynomials:**
Pros:
1. **Stability**: Low-degree polynomials are more stable and robust to variations in the data. They tend to provide simpler and more interpretable models.

2. **Better Generalization**: With lower complexity, low-degree polynomials often generalize better to unseen data. They are less likely to overfit, making them suitable for a wider range of data.

Cons:
1. **High Bias**: Low-degree polynomials have high bias, which means they may underfit the data. They might not capture complex relationships, resulting in model inadequacy.

2. **Limited Representation**: Low-degree polynomials may not be able to accurately represent data with intricate patterns or complex underlying relationships.

The phenomenon at play here is known as the "bias-variance trade-off." In machine learning, the bias-variance trade-off refers to the balance between two sources of error:

1. **Bias**: Bias is the error introduced by approximating a real-world problem with a simplified model. Low-degree polynomials have high bias because they make strong assumptions about the data, leading to underfitting.

2. **Variance**: Variance is the error introduced by the model's sensitivity to fluctuations in the training data. High-degree polynomials have high variance because they are too flexible and sensitive to noise, leading to overfitting.

The goal in machine learning is to find the right balance between bias and variance to create models that generalize well to unseen data. Selecting an appropriate polynomial degree or model complexity is a key decision in achieving this balance. Techniques like cross-validation can help identify the optimal degree for a given problem, minimizing the bias-variance trade-off and building models that generalize effectively.
